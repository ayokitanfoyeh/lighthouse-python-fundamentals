```python
# 200 values from the interval <0,100>, equidistantly divided
x = np.linspace(0,100,200)

# sin(x) values
y1 = np.sin(x)
y2 = (np.sin(x))* (np.cos(x))

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))

# sin(x) plot 
ax.plot(x,y1,'green')

# sin(x)*cos(x) plpt
ax.plot(x,y2,'blue')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (10,6))
axes[0].plot(x,y1,'red')
axes[0].set_title('sin(x)')

# the second plot
axes[1].plot(x,y2,'blue')

# the second plot title
axes[1].set_title('sin(x)*(cos(x))')

# automatically adjust the subplot parameters to give specified padding
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df["target"] = iris.target
```

```python
iris_df['target_name'] = iris_df.target.map(
{0: 'setosa',
 1: 'versicolor',
 2: 'virginica'} 
 )
```

```python
iris_df.head()
```

```python
setosa = iris_df[iris_df.target_name == 'setosa']
versicolor = iris_df[iris_df.target_name == 'versicolor']

# Iris virginica
virginica = iris_df[iris_df.target_name == 'virginica']
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'],
marker = 'o', color ='red', label = 'setosa')
# scatter the plot of versicolor
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')


# scatter the plot of virginica
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')

# set the legend
ax.legend(loc='upper right')

# set the title
ax.set_title('Iris flower')

# set the x-label
ax.set_xlabel('sepal length (cm)')

# set the y-label
ax.set_ylabel('sepal width (cm)')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
ax =plt.axes(projection='3d')
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X,Y = np.meshgrid(x,y)
```

```python
mu_x = 0
variance_x =10
mu_y = 0
variance_y = 10
means = [mu_x, mu_y]
cov = [[variance_x, 0], [0, variance_y]]
mn = multivariate_normal(means, cov)
mn
```

```python
position = np.array([X,Y]).T
Z = mn.pdf(position)
```

```python
fig, ax = plt.subplots(figsize=(12.8,9.8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color ='r')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12.3,10))
ax = plt.axes(projection ='3d')
ax.plot_surface(X, Y, Z, cmap ='jet')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# the surface
ax.contour(X, Y, Z, 40, cmap='jet')
plt.show()
```
