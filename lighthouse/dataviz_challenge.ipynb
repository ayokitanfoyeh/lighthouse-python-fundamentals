## Task
Which teams scored the largest number of goals in FIFA World Cup?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/kitanolowofoyeku/Downloads/results.csv')
data_worldcup = data[(data['tournament'] == 'FIFA World Cup')]

team = data_worldcup.groupby(['home_team']).sum()
team['sum_goals'] = team['home_score'] + team['away_score'] + team['neutral']

max_goals =team['sum_goals'].sort_values(axis=0, ascending = False)
max_goals.index
max10 = (max_goals.nlargest(10)).iloc[::-1]
x = max10.index
y = max10.values
plt.figure(figsize=(20, 20))
plt.barh(x,y)
plt.xlabel('Goals Scored')
plt.ylabel('Nation')

plt.show()
```

## Task
What is the number of matches played in each tournament throughout history?

```python
tournament_matches = data.groupby(['tournament']).size()
x = tournament_matches.index
y = tournament_matches
plt.figure(figsize=(20, 20))
plt.barh(x,y)
plt.xlabel('Tournament')
plt.ylabel('Matches')
plt.show()
```

## Task 
Show the trend in number of matches per year.

```python
data['year'] = pd.DatetimeIndex(data['date']).year
year_matches = data.groupby(['year']).size()
x = year_matches.index
y = year_matches
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Matches')
plt.show()
```
